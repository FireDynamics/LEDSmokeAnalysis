# Dependencies

```python
from ledsa.tools.ledsa2fds import Fds2ledsa, Ledsa2fds
```

# Create FDS DEVC lines from one or more LEDSA datasets

```python
ledsa_sim = Ledsa2fds()

# Load one ore more LEDSA datasets. Give path and dataset name (e.g. camera name)
ledsa_sim.add_experiment_dataset('/path/to/ledsa/dataset/camera_1', 'camera_1')
ledsa_sim.add_experiment_dataset('/path/to/ledsa/dataset/camera_2', 'camera_2')

# Create LEDSA_transmission_devices.txt file containing 'PATH OBSCURATION' devc lines for all camera->LED paths
# File is written to the given path
# You may include the LEDSA_transmission_devices.txt into your FDS file via &CATF OTHER_FILES='LEDSA_transmission_devices.txt'
ledsa_sim.create_transmission_devices_file('/path/to/FDS_simulation')
```

# Create LEDSA dataset of virtual image data from the above FDS devices data

```python
fds_sim = Fds2ledsa()

# Load FDS devc data from the given path (directory)
fds_sim.load_devc_data('/path/to/FDS_simulation')

# Create datasets of virtual image data for LEDSA from the FDS devc data
# Give path (root directory of the original LEDSA dataset where config files are stored) and dataset name (e.g. camera name)
# the virtual image data is written to analysis/channel3 an can be processed via the LESDA analysis step
fds_sim.create_ledsa_dataset('/path/to/ledsa/dataset/camera_1', 'camera_1')
fds_sim.create_ledsa_dataset('/path/to/ledsa/dataset/camera_2', 'camera_2')
```
